```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib
```

```python
dataset = pd.read_csv('/content/drive/MyDrive/ML/Results.csv')
```

```python
sns.set_style("darkgrid")
sns.scatterplot(data=dataset, x="Classifiers", y="Accuracy")
```

```python
sns.set_style("darkgrid")
sns.lineplot(data=dataset, x="Classifiers", y="Accuracy", palette="Blues_d")
```

```python
sns.barplot(data=dataset, x="Classifiers", y="Accuracy", palette="Blues_d")
```

```python
plt.subplot(1, 2, 1)
sns.countplot(x='Classifiers', data=dataset, palette="Blues_d")
plt.subplot(1, 2, 2)
sns.countplot(x='Accuracy', data=dataset)
```

```python
sns.barplot(x="Classifiers", y="Accuracy", data=dataset)
```

```python
sns.histplot(data=dataset, x="Classifiers", binwidth=0.05, hue="Accuracy")
```
